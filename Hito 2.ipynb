```python
import numpy as np
import pandas as pd

df = pd.read_csv('Student_performance.csv')
# print(df.info())

# 1. Tratamiento de nulos.
df['GPA'] = df['GPA'].fillna(df['GPA'].median())

# 2. Eliminación de Outliers
Q1, Q3 = df['Absences'].quantile([0.25, 0.75])
IQR = Q3 - Q1
df = df[~((df['Absences'] < (Q1 - 1.5 * IQR)) | (df['Absences'] > (Q3 + 1.5 * IQR)))]

# 3. Normalización de Strings
ethnicity_map = {0: ' Caucasian ', 1: 'African American', 2: ' Hispanic ', 3: ' Other '}
df['Ethnicity_Name'] = (df['Ethnicity'].map(ethnicity_map)
                        .str.lower().str.strip().str.replace(' ', '_'))

# 4. Feature Engineering
df['StudyTime_Norm'] = (df['StudyTimeWeekly'] - df['StudyTimeWeekly'].min()) / (df['StudyTimeWeekly'].max() - df['StudyTimeWeekly'].min())
df['Engagement_Index'] = (df['StudyTime_Norm'] * 0.7) + ((df['ParentalSupport']/4) * 0.3)

# Conteo de Actividades
df['Participation_Total'] = df[['Sports', 'Music', 'Volunteering', 'Extracurricular']].sum(axis=1)

print(df.head())
```
