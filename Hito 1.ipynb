#  Trabajo Práctico Integrador (TPI)

## Objetivo General

Desarrollar un sistema de análisis y visualización de datos que permita diagnosticar el desempeño
académico, identificar factores de riesgo y proponer soluciones basadas en evidencia. Los
estudiantes deberán demostrar sus habilidades de programación para transformar datos crudos en
un Informe de Gestión Interactivo.


### Hito 1: Adquisición y Planteo

Dataset: Students Performance Dataset

Este dataset contiene información integral de 2.392 estudiantes de secundario, detallando su demografía, hábitos de estudio, la participación de los padres, actividades extracurriculares, y el desempeño académico. 

# Tabla de Contenidos

- Información del Estudiante
- ID del Estudiante
- Datos Demográficos
- Hábitos de Estudio
- Participación de los Padres
- Actividades Extracurriculares
- Rendimiento Académico
- Variable Objetivo: Clase de Calificación

---

## Información del Estudiante

### ID del Estudiante

- **StudentID**: Identificador único asignado a cada estudiante (1001 a 3392).

---

## Datos Demográficos

- **Age**: La edad de los estudiantes varía entre 15 y 18 años.
- **Gender**: Género de los estudiantes, donde:
  - 0: Masculino  
  - 1: Femenino  

- **Ethnicity**: Etnia de los estudiantes, codificada de la siguiente manera:
  - 0: Caucásico  
  - 1: Afroamericano  
  - 2: Asiático  
  - 3: Otro  

- **ParentalEducation**: Nivel educativo de los padres, codificado como:
  - 0: Ninguno  
  - 1: Secundario  
  - 2: Algo de universidad  
  - 3: Licenciatura  
  - 4: Superior  

---

## Hábitos de Estudio

- **StudyTimeWeekly**: Tiempo de estudio semanal en horas (0 a 20).
- **Absences**: Número de ausencias durante el año escolar (0 a 30).
- **Tutoring**: Tutorías:
  - 0: No  
  - 1: Sí  

---

## Participación de los Padres

- **ParentalSupport**: Nivel de apoyo de los padres, codificado como:
  - 0: Ninguno  
  - 1: Bajo  
  - 2: Moderado  
  - 3: Alto  
  - 4: Muy alto  

---

## Actividades Extracurriculares

- **Extracurricular**: Participación en actividades extracurriculares:
  - 0: No  
  - 1: Sí  

- **Sports**: Participación en deportes:
  - 0: No  
  - 1: Sí  

- **Music**: Participación en actividades musicales:
  - 0: No  
  - 1: Sí  

- **Volunteering**: Participación en voluntariado:
  - 0: No  
  - 1: Sí  

---

## Rendimiento Académico

- **GPA**: Promedio de calificaciones (escala de 2.0 a 4.0), influenciado por hábitos de estudio, participación de los padres y actividades extracurriculares.

---

## Variable Objetivo: Clase de Calificación

- **GradeClass**: Clasificación de las calificaciones según el GPA:
  - 0: 'A' (GPA >= 3.5)  
  - 1: 'B' (3.0 <= GPA < 3.5)  
  - 2: 'C' (2.5 <= GPA < 3.0)  
  - 3: 'D' (2.0 <= GPA < 2.5)  
  - 4: 'F' (GPA < 2.0)  

---

## Conclusión

Este dataset ofrece una visión integral de los factores que influyen en el rendimiento académico de los estudiantes, lo que lo hace ideal para investigaciones educativas, desarrollo de modelos predictivos y análisis estadístico.

### Preguntas de alta complejidad

- Pregunta 1: ¿Cómo interactúan el tiempo de estudio, el apoyo parental y la participación en actividades extracurriculares para influir en la probabilidad de que un estudiante alcance un rendimiento alto (A o B)?

- Pregunta 2: ¿Qué combinación de factores (ausencias, bajo apoyo parental y falta de tutorías) predice con mayor precisión la probabilidad de que un estudiante obtenga calificaciones bajas (D o F)?

- Pregunta 3: ¿Existen perfiles de estudiantes para los cuales aumentar el tiempo de estudio no se traduce en una mejora significativa del GPA, y qué variables explican este fenómeno?
